#### Исходные данные

Исходные данные получены в виде Excel файла с 5 листами, для начала посмотрим на данные с помощью средств SQL

Мы видим данные, разбитые по годам, а также есть две дополнительные таблицы: одна содержит данные о питании в отелях, а другая - о скидках.

Посчитав выручку по годам (пока без учета скидок) можно увидеть, что в целом, выручка растет.
Следующим шагом будет выгрузка данных в Power Bi при помощи SQL запроса, чтобы обработка данных в Power Query заняла меньше времени.

Отвечая на поставленные вопросы:

    - Выручка отелей имеет восходящий тренд, хотя и есть спад в 2020 году, что, возможно, связано с карантинными ограничениями из-за ковида
    - Нагрузка на парковки не растет, количество парковочных мест увеличивается, поэтому на данный момент нет необходимости создавать дополнительные парковочные места
    - С конца 2019 года наблюдается снижение количества бронирования номеров, упала выручка, вероятно, это связано с ковидными ограничениями. Однако тренды выручки, средней стоимости номеров, количества броней имеют восходящий характер.
